## Structure of the messages data 

After uploading the chessbuds_messages.json file, I observed a dictionary contains several keys such as participants, title, and messages. The main focus for this task is the messages key, which includes individual messages with details like the sender’s name, timestamp, message content, and reactions. I used pandas.DataFrame() to convert the messages into a DataFrame, where each row represents one message, and columns represent the message details.

### Manipulating the Data

```python
type(messages_df['bumped_message_metadata'].iloc[0])
```

```python
messages_df['bumped_message_metadata'].iloc[0]
```

```python
bump_msgs = [x ['is_bumped'] for x in messages_df['bumped_message_metadata']]
messages_df['bump_msgs'] = bump_msgs
messages_df.head()
```

```python
missing_counts = messages_df.isna().sum()

# Print the missing counts for each column
for column in missing_counts.index:
    print(column + ": " + str(missing_counts[column]))
```

```python
messages_df= messages_df.dropna(subset=['content'])
drop = ['share', 'photos', 'gifs', 'users', 'reactions','bumped_message_metadata']

# Using list comprehension to filter out the columns to keep
New_df = messages_df[[column for column in messages_df.columns if column not in drop]]
New_df
```

## **Row Level Observation**

Each row in the DataFrame represents a single message sent in the conversation. It captures key details such as the **sender_name**, the **content** of the message, and the **timestamp** and we . This structure ensures that each message is an individual observation, with variables like sender, message content, and time(in ms) organized into separate columns.

 ## **Column Level Observations**

**sender_name:** Name of the sender. \
**timestamp_ms:** Timestamp of the message in milliseconds. \
**content:** Content of the message. \
**reactions:** Reactions to the message. \
**type:** Message type (e.g., Share, Generic). \
**is_unsent:** Denotes if the message is unsent. \
**is_taken_down:** Denotes if the message is removed. \
**bumped_message_metadata:** Data about a bumped message. \
**share:** Details about the shared content. \
**photos:** Details about the images that are attached to the message. \
**gifs:** Details on the gifs that are an attachment to the message. \
**users:** Details about the users that are mentioned in the message.he message.

# Describe the final DataFrame meets tidy data principles:

The final DataFrame meets tidy data principles because each variable (e.g., Sender, Message_Content, Timestamp) is organized into its own column, and each observation (a message) forms a single row. Additionally, the DataFrame focuses on a single observational unit—messages—by removing irrelevant columns and also we droped the bumped messages dictonaries and make an another column of having is_bumped false by column nameis **bumped_msgs**.
