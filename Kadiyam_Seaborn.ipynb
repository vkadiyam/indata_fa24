# Seaborn Group 1 Challenge

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic_clean.csv')
```

```python
titanic.head()
```

```python
sns.relplot(
    data= titanic,
    x= "age", y = "fare"
)
plt.show()
```

```python
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5
)
plt.show()
```

```python
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5, hue= 'sex'
)
```

```python
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5, hue='sex', 
    col = 'survived'
)
```

```python
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5, hue='sex', 
    col = 'survived',
    palette = {"male":"blue", "female": "magenta"}
)
```

```python
titanic['Sex'] = titanic['sex'].map({0:"male", 1:"female"})
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5, hue = "sex", 
    col = 'survived',
    palette = {"male":"blue", "female": "magenta"}
)
```

```python
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5, hue = "sex", 
    col = 'survived',
    palette = {"male":"blue", "female": "magenta"},
    size = "pclass"
)
```

```python
titanic['survided_ind'] = titanic['survived'].map({0:"didnotsurvived", 1:"survived"})
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5, hue = "survided_ind", 
    col = 'survived',
    palette = {"didnotsurvived":"#31b4a6", "survived": "#b4316c"}
)
```

```python
sns.relplot(
    data= titanic,
    x= "age", y = "fare",
    alpha = 0.5, hue = "survided_ind", 
    col = 'survived',
    palette = {"didnotsurvived":"#31b4a6", "survived": "#b4316c"},
    size = "pclass"
)
```

# Group 2 Seaborn challenge

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
smh = pd.read_csv('Student Mental health.csv')
smh.head()
```

```python
smh.groupby("Choose your gender")[['Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?']].count()
```

```python
columns_to_plot = smh.columns[-4:-1]
category_order = ['Yes','No']
```

```python
for column in columns_to_plot:
    counts = smh[column].value_counts()
    counts = counts.reindex(category_order, fill_value = 0)
    print(counts)
    print()
```

```python
sns.countplot(smh, x='Do you have Depression?', stat='percent' , hue= 'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x='Do you have Anxiety?', stat='percent' , hue= 'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x='Do you have Panic attack?', stat='percent' , hue= 'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x='Do you have Depression?', stat='percent' , hue= 'Choose your gender', palette ='PuBuGn')
```

```python
sns.countplot(smh, x='Do you have Anxiety?', stat='percent' , hue= 'Choose your gender', palette = 'PuBu')
```
